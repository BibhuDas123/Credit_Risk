**OUTLIER CLASS**

```python
#Let us do an outler analysis
class OutlierAnalysis(object):
    '''Classes with different functions which helps in outlier analysis'''
    
    def __init__(self, name):
        """Return a Customer object whose name is *name*.""" 
        self.name = name
    
    def diagnostic_plots(self,df, variable):
        '''function takes a dataframe (df) and the variable of interest as arguments'''

        # define figure size
        plt.figure(figsize=(16, 4))

        # histogram
        plt.subplot(1, 3, 1)
        sns.distplot(df[variable], bins=40)
        plt.title('Histogram')

        # Q-Q plot
        plt.subplot(1, 3, 2)
        stats.probplot(df[variable], dist="norm", plot=plt)
        plt.ylabel('Variable quantiles')

        # boxplot
        plt.subplot(1, 3, 3)
        sns.boxplot(y=df[variable])
        plt.title('Boxplot')

        plt.show()
    
    def find_skewed_boundaries(self,df, variable, distance):

        '''Let's calculate the boundaries outside which sit the outliers for skewed distributions.distance passed as an argument, gives us the option to estimate 1.5 times or 3 times the IQR to calculate the boundaries.'''

        IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

        lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
        upper_boundary = df[variable].quantile(0.75) + (IQR * distance)

        return upper_boundary, lower_boundary

    def capper_skewed(self,data,input_param):
        windsoriser = Winsorizer(distribution='skewed', # choose skewed for IQR rule boundaries or gaussian for mean and std
                          tail='both', # cap left, right or both tails 
                          fold=1.5,
                          variables=input_param)

        data=windsoriser.fit_transform(data)
        return data
        
    
    def find_boundaries(self,df, variable):

        ''' the boundaries are the quantiles '''

        lower_boundary = df[variable].quantile(0.05)
        upper_boundary = df[variable].quantile(0.95)

        return upper_boundary, lower_boundary
    
    def capper_quantiles(self,data,input_param):
        windsoriser = Winsorizer(distribution='quantiles', # choose skewed for IQR rule boundaries or gaussian for mean and std
                          tail='both', # cap left, right or both tails 
                          fold=1.5,
                          variables=input_param)

        data=windsoriser.fit_transform(data)
        return data
        
    def find_normal_boundaries(self,df, variable, distance):

        '''calculate the boundaries outside which sit the outliers for a Gaussian distribution'''

        upper_boundary = df[variable].mean() + distance * df[variable].std()
        lower_boundary = df[variable].mean() - distance * df[variable].std()

        return upper_boundary, lower_boundary

    def capper_gaussian(self,data,input_param):
        windsoriser = Winsorizer(distribution='gaussian', # choose skewed for IQR rule boundaries or gaussian for mean and std
                          tail='both', # cap left, right or both tails 
                          fold=1.5,
                          variables=input_param)

        data=windsoriser.fit_transform(data)
        return data
    
```

```python
#Feature selection class
class FeatureSelection(object):
    '''This class represents the steps to be taken in the Feature Selection '''
    
    def __init__(self, name):
        """Return a Customer object whose name is *name*.""" 
        self.name = name
        
    def variancethreshold(self,X_train,X_test):
        '''It finds the columns with are constant and provides the output in the form of number of features that 
        are not constant '''
        if X_train.isnull().sum()==0 and X_test.isnull().sum()==0:
            sel=VarianceThreshold(threshold=0).fit(X_train)
            constant=sum(sel.get_support())
            print([x for x in X_train.columns if x  not in X_train.columns[sel.get_support()]])
            X_train = sel.transform(X_train)
            X_test = sel.transform(X_test)
            return X_train,X_test
        else:
            print('The training set has missing values')
            
    def varisncethreshold_std(self,X_train,X_test):
        '''Using the to remove columns with constant variables'''
        if X_train.isnull().sum()==0 and X_test.isnull().sum()==0:
            constant_features=[x for x in X_train.columns if X_train.columns[feat].std()==0]
            X_train.drop(labels=constant_features, axis=1, inplace=True)
            X_test.drop(labels=constant_features, axis=1, inplace=True)
            print(X_train.shape, X_test.shape)
            return X_train,X_test
        else:
            print('The training set has missing values')
            
    def variancethreshold_quasi(self,X_train,X_test):
        '''Variance threshold from sklearn is a simple baseline approach to feature selection. It removes all features which variance doesn’t meet some threshold. By default, it removes all zero-variance features, i.e., features that have the same value in all samples. '''
        if X_train.isnull().sum()==0 and X_test.isnull().sum()==0:
            sel=VarianceThreshold(threshold=0.01).fit(X_train)
            constant=sum(sel.get_support())
            print([x for x in X_train.columns if x  not in X_train.columns[sel.get_support()]])
            X_train = sel.transform(X_train)
            X_test = sel.transform(X_test)
            return X_train,X_test
        else:
            print('The training set has missing values')
            
    def remove_duplicates(self,X_train,X_test):
        '''Removing duplicate features'''
        if X_train.isnull().sum()==0 and X_test.isnull().sum()==0:
            X_train_t, X_test_t=X_train.T, X_test.T
            print(X_train_t.duplicated.sum())
            duplicated_features = X_train_t[X_train_t.duplicated()].index.values
            X_train=X_train_t.duplicated(keep='first').T
            X_test = X_test_t.duplicated(keep='first').T
            return X_train,X_test
        else:
            print('The training set has missing values')
    
    def correlation(self,X_train,X_test,threshold):
        '''Removal of columns which are highly correlated'''
        col_corr=set()
        corr_matrix=dataset.corr()
        for i in range(corr_matrix.columns):
            for j in range(i):
                if abs(corr_matrix.iloc[i,j])>threshold:
                    colname=corr_matrix.columns[i]
                    col_corr.add(colname)
        X_train.drop(labels=col_corr,axis=1,inplace=True)
        X_test.drop(labels=col_corr,axis=1,inplace=True)
        return X_train,X_test
    
    def corrmat(self,X_train):
        '''Finding the features which are highly correalted using correlation matrix'''
        corrmat=X_train.corr().abs().unstack().sort_values(ascending=False)
        corrmat=corrmat[(corrmat>=0.8)& (corrmat<1)]
        corrmat=pd.DataFrame(corrmat).reset_index()
        corrmat.columns = ['feature1', 'feature2', 'corr']
        grouped_feature_ls=[]
        correlated_features=[]
        for feature in corrmat.feature1.unique():
            if feature not in grouped_feature_ls:
                correlated_block=corrmat[corrmat.feature1==feature]
                grouped_feature_ls=grouped_feature_ls +list(correlated_block.feature2.unique())+[feature]
                correlated_features.append(correlated_block)
        print('found {} correlated groups'.format(len(correlated_features)))
        print('out of {} total features'.format(X_train.shape[1]))
        return correlated_features
    
    def feature_importance_rf():
        '''Random forest feature importance '''
        rf=RandomForestClassifier(n_estimators=200, random_state=39, max_depth=4)
        rf.fit(X_train,y_train)
        features=list(X_train.columns)
        importance = pd.concat(
                [pd.Series(features),
             pd.Series(rf.feature_importances_)], axis=1)

        importance.columns = ['feature', 'importance']
        importance.sort_values(by='importance', ascending=False)
        return importance
    
    def mutual_info_classification(self,X_train,y_train):
        '''calculate the mutual information between the variables and the target this returns the mutual information value of each feature
        The smaller the value the less information the feature has about the target'''
        
        mi=pd.Series(mutual_info_classif(X_train.fillna(0),y_train))
        mi.index=X_train.columns
        mi.sort_values(ascending=False).plot.bar(figsize=(20,8))
        sel=SelectKBest(mutual_info_classif,k=10).fit(X_train,y_train)
        X_train=X_train.columns[sel.get_support()]
        return X_train
    
    def mutual_info_regression(self,X_train,y_train):
        '''calculate the mutual regression between the variables and the target this returns the mutual information value of each feature
        The smaller the value the less information the feature has about the target'''
        mi=pd.Series(mutual_info_regression(X_train.fillna(0),y_train))
        mi.index=X_train.columns
        mi.sort_values(ascending=False).plot.bar(figsize=(20,8))
        sel = SelectPercentile(mutual_info_regression, percentile=10).fit(X_train.fillna(0), y_train)
        X_train=X_train.columns[sel_.get_support()]
        return X_train
    
    def fischer_score(self,X_train,X_test):
        ''' For Fisher score(ONLY APPLIED ON CATEGORICAL VARIABLES),the smaller the p_value, the more significant the feature is to predict the target
        Note: One thing to keep in mind when using Fisher score or univariate selection methods, is that in very big datasets, 
        most of the features will show a small p_value, and therefore look like they are highly predictive. 
        This is in fact an effect of the sample size. So care should be taken when selecting features using these procedures. 
        An ultra tiny p_value does not highlight an ultra-important feature, it rather indicates that the dataset contains 
        too many samples.Finally, in this demonstration, we could then combine this procedure with SelectKBest or SelectPercentile,'''
        f_score=chi(X_train.fillna(0),y_train)
        pvalues=pd.Series(f_score[1])
        pvalues.index=X_train.columns
        pvalues=pvalues.sort_values(ascending=False)
        return pvalues
    
    def univariate_selection_classfication(self,X_train,X_test):
        '''Univariate feature selection works by selecting the best features based on univariate statistical tests (ANOVA). 
        The methods based on F-test estimate the degree of linear dependency between two random variables. They assume 
        a linear relationship between the feature and the target. These methods also assume that the variables follow a 
        Gaussian distribution.Remember that the lower the p_value, the most predictive the feature is in principle. There 
        are a few features that do not seem to have predictive power according to this tests, which are those on the left 
        with pvalues above 0.05. Given that typically in statistics one uses a pvalue of 0.05 as a cut-off, I am inclined 
        to believe that those features with pvalue > 0.05 are indeed not important. However, keep in mind that this test 
        assumes a linear relationship, so it might also be the case that the feature is related to the target but not in 
        a linear manner.in big datasets it is not unusual that the pvalues of the different features are really small. This
        does not say as much about the relevance of the feature. Mostly it indicates that it is a big the dataset.'''
        
        univariate=f_classif(X_train.fillna(0),y_train)
        univariate=ps.Series(univariate[1])
        univariate.index=X_train.columns
        univariate=univariate.sort_values(ascending=False).plot.bar(figsize=(20, 8))
        sel = SelectKBest(f_classif, k=10).fit(X_train.fillna(0), y_train)
        X_train=X_train.columns[sel.get_support()]
        return X_train
    
    def univariate_selection_regression(self,X_train,X_test):
        '''Univariate feature selection works by selecting the best features based on univariate statistical tests (ANOVA). 
        The methods based on F-test estimate the degree of linear dependency between two random variables. They assume 
        a linear relationship between the feature and the target. These methods also assume that the variables follow a 
        Gaussian distribution.Remember that the lower the p_value, the most predictive the feature is in principle. There 
        are a few features that do not seem to have predictive power according to this tests, which are those on the left 
        with pvalues above 0.05. Given that typically in statistics one uses a pvalue of 0.05 as a cut-off, I am inclined 
        to believe that those features with pvalue > 0.05 are indeed not important. However, keep in mind that this test 
        assumes a linear relationship, so it might also be the case that the feature is related to the target but not in 
        a linear manner.in big datasets it is not unusual that the pvalues of the different features are really small. This
        does not say as much about the relevance of the feature. Mostly it indicates that it is a big the dataset.'''
        
        univariate=f_regression(X_train.fillna(0),y_train)
        univariate=ps.Series(univariate[1])
        univariate.index=X_train.columns
        univariate=univariate.sort_values(ascending=False).plot.bar(figsize=(20, 8))
        sel=SelectPercentile(f_regression,percentile=10).fit(X-train.fillna(0),y_train)
        X_train=X_train.columns[sel.get_support()]
        return X_train
    
    
    def recursive_feature_selection(self,X_train,X_test):
        '''Random Forests assign equal or similar importance to features that are highly correlated. In addition, when features are correlated, the importance assigned is 
        lower than the importance attributed to the feature itself, should the tree be built without the correlated counterparts.The cycle is as follows:

            1.Build random forests using all features
            2.Remove least important feature
            3.Build random forests and recalculate importance
            4.Repeat until a criteria is met'''

        sel=RFE(RandomForestClassifier(n_estimators=100),n_features_to_select=10)
        sel.fit(X_train.fillna(0),y_train)
        selected_feat = X_train.columns[(sel_.get_support())]
        return selected_feat
    
    
    def Lasso(self,X_train,y_train):
        '''Regularisation consists in adding a penalty to the different parameters of the machine learning model to reduce the freedom of the 
        model and in other words to avoid overfitting. In linear model regularisation, the penalty is applied over the coefficients that multiply 
        each of the predictors. From the different types of regularisation, Lasso or l1 has the property that is able to shrink some of the 
        coefficients to zero. Therefore, that feature can be removed from the model.'''
        
        scaler = StandardScaler()
        scaler.fit(X_train.fillna(0))
        sel = SelectFromModel(LogisticRegression(C=1, penalty='l2'))
        sel.fit(scaler.transform(X_train.fillna(0)), y_train)
        selected_features=X_train.columns[sel.get_support()]
        print('total_features:{}'.format(X_train.shape[1]))
        print('selected_features:{}'.format(len(selected_features)))
        print('features with zero coefficients:{}'.format(np.sum(sel.estimator_.coef_==0)))
       
    def LassoCV(self,X_train,Y_train):
        '''Using lasso regression technique to find out the redundant features in the equation'''
        
        scaler = StandardScaler()
        scaler.fit(X_train_c.fillna(0))
        sel = LassoCV()
        sel.fit(scaler.transform(X_train_c.fillna(0)), y_train)
        coef = pd.Series(sel.coef_, index = X_train.columns)
        print('total_features:{}'.format(X_train.shape[1]))
        print('selected_features:{}'.format(str(sum(coef != 0))))
        print('features with zero coefficients:{}'.format(np.sum(sel.coef_==0)))
        imp_coef = coef.sort_values()
        matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
        imp_coef.plot(kind = "barh")
        plt.title("Feature importance using Lasso Model")
        
        
    
    def recursive_feature_addition_classification(self,X_train,X_test,y_train,y_pred):
        model_all_features=xgb.XGBClassifier(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
        model_all_features.fit(X_train,y_train)
        y_pred_test=model_all_features.predict(X_test)[:,1]
        auc_score_all=roc_auc_score(y_pred,y_pred_test)
        print('AUC_ROC score of all features=%f'%(auc_score_all))

        features=pd.Series(model_all_features.feature_importances_)
        features.index=X_train.columns
        features.sort_values(ascending=False).plot.bar(figsize(20,8))
        features=list(features.index)

        #Now taking one by one important feature
        model_one_feature=xgb.XGBClassifier(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
        model_one_feature.fit(X_train[features[0]].to_frame(),y_train)
        y_pred_test=model_one_feature.predict(X_test[features[0]].to_frame())[:,1]
        auc_score_one=roc_auc_score(y_pred,y_pred_test)
        print('AUC_ROC score of all features=%f'%(auc_score_one))

        # the final step consists in adding one at a time
        # all the features, from the most to the least
        # important, and build an xgboost at each round.

        # once we build the model, we calculate the new roc-auc
        # if the new roc-auc is bigger than the original one
        # (with one feature), then that feature that was added
        # was important, and we should keep it.
        # otherwise, we should remove the feature

        # recursive feature addition:

        # first we arbitrarily set the increase in roc-auc
        # if the increase is above this threshold,
        # the feature will be kept
        tol = 0.001

        features_to_keep=[features[0]]
        count=1
        for feature in features[1:]:
            print('testing feature',feature,' which is feature ',count,' out of ', len(features))
            count=count+1
            model_int = xgb.XGBClassifier(
                nthread=10, max_depth=4, n_estimators=500, learning_rate=0.05)

            # fit model with the selected features
            # and the feature to be evaluated
            model_int.fit(
                X_train[features_to_keep + [feature] ], y_train)

            # make a prediction over the test set
            y_pred_test = model_int.predict_proba(
                X_test[features_to_keep + [feature] ])[:, 1]

            # calculate the new roc-auc
            auc_score_int = roc_auc_score(y_test, y_pred_test)
            print('New Test ROC AUC={}'.format((auc_score_int)))

            # print the original roc-auc with one feature
            print('All features Test ROC AUC={}'.format((auc_score_one)))

            # determine the increase in the roc-auc
            diff_auc = auc_score_int - auc_score_one

            if diff_auc >= tol:
                print('Increase in ROC AUC={}'.format(diff_auc))
                print('keep: ', feature)
                print
                auc_score_one=auc_score_int
                features_to_keep.append(feature)
            else:
                print('Increase in ROC AUC={}'.format(diff_auc))
                print('remove: ', feature)
                print
        print('DONE!!')
        print('total features to keep: ', len(features_to_keep))
        return features_to_keep

    def recursive_feature_addition_regression(self,X_train,y_train,X_test,y_test):
        model_all_features=xgb.XGBRegressor(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
        model_all_features.fit(X_train,y_train)
        y_pred_test=model_all_features.predict(X_test)
        r2_score_all=r2_score(y_test,y_pred_test)
        print('R2 score of all features=%f'%(r2_score_all))

        features=pd.Series(model_all_features.feature_importances_)
        features.index=X_train.columns
        features.sort_values(ascending=False).plot.bar(figsize=(20,8))
        features=list(features.index)

            #Now taking one by one important feature
        model_one_feature=xgb.XGBRegressor(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
        model_one_feature.fit(X_train[features[0]].to_frame(),y_train)
        y_pred_test=model_one_feature.predict(X_test[features[0]].to_frame())
        r2_score_one=r2_score(y_test,y_pred_test)
        print('R2 score of all features=%f'%(r2_score_one))

        tol = 0.001

        features_to_keep=[features[0]]
        count=1
         # the final step consists in adding one at a time
        # all the features, from the most to the least
        # important, and build an xgboost at each round.

        # once we build the model, we calculate the new roc-auc
        # if the new roc-auc is bigger than the original one
        # (with one feature), then that feature that was added
        # was important, and we should keep it.
        # otherwise, we should remove the feature

        # recursive feature addition:

        # first we arbitrarily set the increase in roc-auc
        # if the increase is above this threshold,
        # the feature will be kept
        for feature in features[1:]:
            print('testing feature',feature,' which is feature ',count,' out of ', len(features))
            count=count+1
            model_int = xgb.XGBRegressor(
                    nthread=10, max_depth=4, n_estimators=500, learning_rate=0.05)

                # fit model with the selected features
                # and the feature to be evaluated
            model_int.fit(
                    X_train[features_to_keep + [feature] ], y_train)

                # make a prediction over the test set
            y_pred_test = model_int.predict(
                    X_test[features_to_keep + [feature] ])

                # calculate the new roc-auc
            r2_score_int = r2_score(y_test, y_pred_test)
            print('New Test R2_score={}'.format((r2_score_int)))

                # print the original roc-auc with one feature
            print('All features Test R2_score={}'.format((r2_score_one)))

                # determine the increase in the roc-auc
            diff_r2 = r2_score_int - r2_score_one

            if diff_r2 >= tol:
                print('Increase in R2={}'.format(diff_r2))
                print('keep: ', feature)
                print
                r2_score_one=r2_score_int
                features_to_keep.append(feature)
            else:
                print('Increase in R2={}'.format(diff_r2))
                print('remove: ', feature)
                print
            print('DONE!!')
            print('total features to keep: ', len(features_to_keep))
        return features_to_keep

   
        def recursive_feature_elimination_classification(self,X_train,X_test,y_train,y_pred):
            model_all_features=xgb.XGBClassifier(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
            model_all_features.fit(X_train,y_train)
            y_pred_test=model_all_features.predict(X_test)[:,1]
            auc_score_all=roc_auc_score(y_pred,y_pred_test)
            print('AUC_ROC score of all features=%f'%(auc_score_all))

            features=pd.Series(model_all_features.feature_importances_)
            features.index=X_train.columns
            features.sort_values(ascending=False).plot.bar(figsize(20,8))
            features=list(features.index)

            #Now taking one by one important feature
            model_one_feature=xgb.XGBClassifier(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
            model_one_feature.fit(X_train[features[0]].to_frame(),y_train)
            y_pred_test=model_one_feature.predict(X_test[features[0]].to_frame())[:,1]
            auc_score_one=roc_auc_score(y_pred,y_pred_test)
            print('AUC_ROC score of all features=%f'%(auc_score_one))



            # recursive feature elimination:

            # the first step of this procedure  consists in building
            # a machine learning algorithm using all the available features
            # and then determine the importance of the features according
            # to the algorithm

            # the second step consist of deriving the importance of 
            # each feature and ranking them from the least to the most
            # important

            # the final step consists in removing one at a time
            # all the features, from the least to the most
            # important, and build an xgboost at each round.

            # once we build the model, we calculate the new roc-auc
            # if the new roc-auc is smaller than the original one
            # (with all the features), then that feature that was removed
            # was important, and we should keep it.
            # otherwise, we should remove the feature

            tol = 0.0001

            features_to_remove=[]
            count=1
            for feature in features:
                print('testing feature',feature,' which is feature ',count,' out of ', len(features))
                count=count+1
                model_int = xgb.XGBClassifier(
                    nthread=10, max_depth=4, n_estimators=500, learning_rate=0.05)

                # fit model with the selected features
                # and the feature to be evaluated
                model_int.fit(
                    X_train.drop(features_to_remove + [feature],axis=1), y_train)

                # make a prediction over the test set
                y_pred_test = model_int.predict_proba(
                    X_test.drop(features_to_keep + [feature],axis=1))[:, 1]

                # calculate the new roc-auc
                auc_score_int = roc_auc_score(y_test, y_pred_test)
                print('New Test ROC AUC={}'.format((auc_score_int)))

                # print the original roc-auc with one feature
                print('All features Test ROC AUC={}'.format((auc_score_one)))

                # determine the increase in the roc-auc
                diff_auc = auc_score_int - auc_score_one

                if diff_auc >= tol:
                    print('Drop in ROC AUC={}'.format(diff_auc))
                    print('keep: ', feature)
                    print

                else:
                    print('Drop in ROC AUC={}'.format(diff_auc))
                    print('remove: ', feature)
                    print
                    auc_score_one=auc_score_int
                    features_to_remove.append(feature)
            print('DONE!!')
            print('total features to remove: ', len(features_to_remove))
        return features_to_remove
    
    
    def recursive_feature_addition_regression(self,X_train,y_train,X_test,y_test):
        model_all_features=xgb.XGBRegressor(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
        model_all_features.fit(X_train,y_train)
        y_pred_test=model_all_features.predict(X_test)
        r2_score_all=r2_score(y_test,y_pred_test)
        print('R2 score of all features=%f'%(r2_score_all))

        features=pd.Series(model_all_features.feature_importances_)
        features.index=X_train.columns
        features.sort_values(ascending=False).plot.bar(figsize=(20,8))
        features=list(features.index)

            #Now taking one by one important feature
        model_one_feature=xgb.XGBRegressor(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
        model_one_feature.fit(X_train[features[0]].to_frame(),y_train)
        y_pred_test=model_one_feature.predict(X_test[features[0]].to_frame())
        r2_score_one=r2_score(y_test,y_pred_test)
        print('R2 score of all features=%f'%(r2_score_one))

        tol = 0.0001

        features_to_remove=[]
        count=1
        # the first step of this procedure  consists in building
        # a machine learning algorithm using all the available features
        # and then determine the importance of the features according
        # to the algorithm

        # the second step consist of deriving the importance of 
        # each feature and ranking them from the least to the most
        # important

        # the final step consists in removing one at a time
        # all the features, from the least to the most
        # important, and build an xgboost at each round.

        # once we build the model, we calculate the new roc-auc
        # if the new roc-auc is smaller than the original one
        # (with all the features), then that feature that was removed
        # was important, and we should keep it.
        # otherwise, we should remove the feature

        for feature in features:
            print('testing feature',feature,' which is feature ',count,' out of ', len(features))
            count=count+1
            model_int = xgb.XGBRegressor(
                    nthread=10, max_depth=4, n_estimators=500, learning_rate=0.05)

                # fit model with the selected features
                # and the feature to be evaluated
            model_int.fit(
                    X_train.drop(features_to_remove + [feature],axis=1), y_train)

                # make a prediction over the test set
            y_pred_test = model_int.predict(
                    X_test.drop(features_to_remove + [feature],axis=1))

                # calculate the new roc-auc
            r2_score_int = r2_score(y_test, y_pred_test)
            print('New Test R2_score={}'.format((r2_score_int)))

                # print the original roc-auc with one feature
            print('All features Test R2_score={}'.format((r2_score_one)))

                # determine the increase in the roc-auc
            diff_r2 = r2_score_int - r2_score_one

            if diff_r2 >= tol:
                print('Increase in R2={}'.format(diff_r2))
                print('keep: ', feature)
                print

            else:
                print('Increase in R2={}'.format(diff_r2))
                print('remove: ', feature)
                print
                r2_score_one=r2_score_int
                features_to_remove.append(feature)
            print('DONE!!')
            print('total features to keep: ', len(features_to_remove))
        return features_to_remove
    
    def recursive_feature_elimination_regression(self,X_train_c,X_test_c,y_train_c,y_test_c):
        
            model_all_features = xgb.XGBRegressor(
                nthread=10, max_depth=4, n_estimators=500, learning_rate=0.05,random_state=123)

            model_all_features.fit(X_train_c, y_train_c)

            # calculate the roc-auc in the test set
            y_pred_test = model_all_features.predict(X_test_c)
            r2_score_all = r2_score(y_test_c, y_pred_test)
            print('Test all features xgb R2 = %f' % (r2_score_all))

            features = pd.Series(model_all_features.feature_importances_)
            features.index = X_train_c.columns

            # sort the features by importance
            features.sort_values(ascending=True, inplace=True)

            # plot
            features.plot.bar(figsize=(20,6))
            features = list(features.index)

            tol = 0.001

            print('doing recursive feature elimination')

            # we initialise a list where we will collect the
            # features we should remove
            features_to_remove = []

            # set a counter to know how far ahead the loop is going
            count = 1

            # now we loop over all the features, in order of importance:
            # remember that features is the list of ordered features
            # by importance
            for feature in features:
                print()
                print('testing feature: ', feature, ' which is feature ', count,
                      ' out of ', len(features))
                count = count + 1

                # initialise model
                model_int = xgb.XGBRegressor(
                    nthread=10, max_depth=4, n_estimators=500, learning_rate=0.05,random_state=123)

                # fit model with all variables minus the removed features
                # and the feature to be evaluated
                model_int.fit(
                    X_train_c.drop(features_to_remove + [feature], axis=1), y_train_c)

                # make a prediction over the test set
                y_pred_test = model_int.predict(
                    X_test_c.drop(features_to_remove + [feature], axis=1))

                # calculate the new r2
                r2_score_int = r2_score(y_test_c, y_pred_test)
                print('New Test r2 = {}'.format((r2_score_int)))

                # print the original r2 with all the features
                print('All features Test r2 = {}'.format((r2_score_all)))

                # determine the drop in the r2
                diff_r2 = r2_score_all - r2_score_int

                # compare the drop in r2 with the tolerance
                # we set previously
                if diff_r2 >= tol:
                    print('Drop in r2 ={}'.format(diff_r2))
                    print('keep: ', feature)
                    print
                else:
                    print('Drop in r2 = {}'.format(diff_r2))
                    print('remove: ', feature)
                    print
                    # if the drop in the r2 is small and we remove the
                    # feature, we need to set the new r2 to the one based on
                    # the remaining features
                    r2_score_all = r2_score_int

                    # and append the feature to remove to the collecting
                    # list
                    features_to_remove.append(feature)

            # now the loop is finished, we evaluated all the features
            print('DONE!!')
            print('total features to remove: ', len(features_to_remove))

            # determine the features to keep (those we won't remove)
            features_to_keep = [x for x in features if x not in features_to_remove]
            print('total features to keep: ', len(features_to_keep))
            return features_to_keep

    
    def feature_shuffling_regression(self,X_train,y_train,X_test,y_test):
        model=xgb.XGBRegressor(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
        model.fit(X_train,y_train)
        y_pred_test=model.predict(X_test)
        r2_score_all=r2_score(y_test,y_pred_test)
        print('R2 score of all features=%f'%(r2_score_all))

        print('train r2 score:',r2_score(y_train,(model.predict(X_train))))
        print('test r2 score:',r2_score(y_test,(model.predict(X_test))))

        train_r2=r2_score(y_train,model.predict(X_train))

        feature_dict={}

        for feature in X_train.columns:
            X_train_c=X_train.copy()
            X_train_c[feature] = X_train_c[feature].sample(frac=1).reset_index(
                drop=True)

            # make prediction with shuffled feature and calculate roc-auc
            shuff_r2 = r2_score(y_train,
                                      (model.predict(X_train_c.fillna(0))))

            # save the drop in roc-aucsel
            feature_dict[feature] = (train_r2 - shuff_r2)

        feature_importance = pd.Series(feature_dict).reset_index()
        feature_importance.columns = ['feature', 'auc_drop']
        feature_importance.sort_values(by=['auc_drop'], ascending=False, inplace=True)
        feature_importance.head()
        temp = pd.Series(feature_importance[feature_importance.auc_drop>0]['auc_drop'])
        temp.index = pd.Series(feature_importance[feature_importance.auc_drop>0]['feature'])

        pd.Series(temp).plot.bar(figsize=(15,6))
        selected_features = feature_importance[feature_importance.auc_drop>0]['feature']
        return selected_features
    
    
    
    def feature_shuffling_classification(self,X_train,y_train,X_test,y_test):
        model=xgb.XGBClassifier(nthread=10,n_estimators=500,learning_rate=0.01,max_depth=4,random_state=123)
        model.fit(X_train,y_train)
        y_pred_test=model.predict(X_test)
        auc_roc_score_all=auc_roc_score(y_test,y_pred_test)
        print('AUC_ROC score of all features=%f'%(auc_roc_score_all))

        print('train auc_roc score:',auc_roc_score(y_train,(model.predict_proba(X_train))[:,1]))
        print('test auc_roc score:',auc_roc_score(y_test,(model.predict_proba(X_test))[:,1]))

        train_auc_roc=auc_roc_score(y_train,model.predict_proba(X_train)[:,1])

        feature_dict={}

        for feature in X_train.columns:
            X_train_c=X_train.copy()
            X_train_c[feature] = X_train_c[feature].sample(frac=1).reset_index(
                drop=True)

            # make prediction with shuffled feature and calculate roc-auc
            shuff_auc_roc = auc_roc_score(y_train,
                                      (model.predict_proba(X_train_c.fillna(0)))[:,1])

            # save the drop in roc-auc
            feature_dict[feature] = (train_auc_roc - shuff_auc_roc)

        feature_importance = pd.Series(feature_dict).reset_index()
        feature_importance.columns = ['feature', 'auc_drop']
        feature_importance.sort_values(by=['auc_drop'], ascending=False, inplace=True)
        feature_importance.head()
        temp = pd.Series(feature_importance[feature_importance.auc_drop>0]['auc_drop'])
        temp.index = pd.Series(feature_importance[feature_importance.auc_drop>0]['feature'])

        pd.Series(temp).plot.bar(figsize=(15,6))
        selected_features = feature_importance[feature_importance.auc_drop>0]['feature']
        return selected_features
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
def metrics_error(y_true,y_pred):
    # calculate MAE, MSE, RMSE
    print('R2 score : '+str(r2_score(y_true, y_pred)))
    print('mean absolute error : '+str(metrics.mean_absolute_error(y_true, y_pred)))
    print('mean squared error : '+str(metrics.mean_squared_error(y_true, y_pred)))
    print('root mean squared error : '+str(np.sqrt(metrics.mean_squared_error(y_true, y_pred))))
```

**GRID SEARCH**

```python
from sklearn.model_selection import GridSearchCV
def grid_search_algo(X_train,y_train,X_test,y_test,model,param_grid,cv=5):
    gs=GridSearchCV(estimator=model,param_grid=param_grid,cv=cv,n_jobs=-1,verbose=1)
    grid=gs.fit(X_train,y_train)
    y_pred_train=gs.predict(X_train)
    y_pred_test=gs.predict(X_test)
    print(' R-squared:',gs.score(X_test, y_test))
    print("The model performance for training set")
    print("--------------------------------------")
    metrics_error(y_test,y_pred_train)
    print("The model performance for testing set")
    print("--------------------------------------")
    metrics_error(y_test,y_pred_test)
    return gs
    
```
