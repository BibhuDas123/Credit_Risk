```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn import decomposition
import matplotlib.pyplot as plt
import numpy as np
import re,nltk
from nltk.stem.porter import PorterStemmer
from sklearn.model_selection import train_test_split
```

```
nltk.download('punkt')
```

```
df=pd.read_csv('https://github.com/srivatsan88/YouTubeLI/blob/master/dataset/consumer_compliants.zip?raw=true',compression='zip',sep=',',quotechar='"')
```

```
df['Product'].value_counts()
```

```
df['Company'].value_counts()
```

```
complaints_df=df[['Consumer complaint narrative','Product','Company']].rename(columns={'Consumer complaint narrative':'complaints'})
pd.set_option('display.max_colwidth',-1)
complaints_df
```

```
X_train,X_test=train_test_split(complaints_df,test_size=0.6,random_state=0)
```

```
stemmer=PorterStemmer()
nltk.download('stopwords')
stop_words=set(nltk.corpus.stopwords.words('english'))
```

```
def tokenize(text):
  tokens=[word for word in nltk.word_tokenize(text) if (len(word)>3 and len(word.strip('Xx/'))>2)]
  stems=[stemmer.stem(item) for item in tokens]
  return stems
```

```
#Intailsization the tfidf to convert words to vector format.It is used for LDA as it is used to find count of particular word in the document to create similarity between words.When we use idf as fasle then idf is not exceuted it only exceutes tf(term frequency)
vectorizer_tf=TfidfVectorizer(tokenizer=tokenize,stop_words='english',max_df=0.75,min_df=50,max_features=10000,use_idf=False,norm=None)
tf_vectors=vectorizer_tf.fit_transform(X_train.complaints)
```

```
tf_vectors.A
```

```
vectorizer_tf.get_feature_names()
```

```
#LDA is a statistical model that allow s a set of observations aby a group or unobserved group.There are two learning method batch and online.Batch uses all the the words at once but online method is more prefeably for large dataset  and it runs a mini batch dataframe and learning offset tells us how you are going to increase the no of words with each online mini dataframe 
lda=decomposition.LatentDirichletAllocation(n_components=6,max_iter=3,learning_method='online',learning_offset=50,n_jobs=-1,random_state=0)
W1=lda.fit_transform(tf_vectors)
H1=lda.components_
```

```
W1
```

```
#Try to identify the top words whast the topic the customewr is talking about
num_words=15
vocab=np.array(vectorizer_tf.get_feature_names())

top_words=lambda t:[vocab[i] for i in np.argsort(t)[:-num_words-1:-1]]
topic_words=([top_words(t) for t in H1])
topics=[" ".join(t) for t in topic_words]
```

```
topics
```

```
colnames=['Topic'+ str(i) for i in range(lda.n_components)]
docnames=['Doc'+str(i) for i in range(len(X_train.complaints))]
df_doc_topic=pd.DataFrame(np.round(W1,2),columns=colnames,index=docnames)
significant_topic=np.argmax(df_doc_topic.values,axis=1)
df_doc_topic['dominant_topic']=significant_topic
df_doc_topic
```
